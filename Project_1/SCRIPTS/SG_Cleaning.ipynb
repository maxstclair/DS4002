```python
import pandas as pd
```

Black Cow

```python
bc_df = pd.read_csv('/content/DS4002/Project_1/data/blackcow_reviews.csv')
```

```python
bc_df['Date'] = pd.to_datetime(bc_df['Date']).dt.date
bc_df['Restaurant'] = 'Black Cow'
bc_df = bc_df.drop(columns=['Author','Cool','Funny','Useful','Author Friends'], axis=1)
bc_df.head()
```

```python
bang_df = pd.read_csv('/content/DS4002/Project_1/data/bang_reviews.csv')
```

Orzo

```python
o_df = pd.read_csv('/content/DS4002/Project_1/data/orzo.csv')
```

```python
o_df['Restaurant'] = 'Orzo'
o_df = o_df.drop(columns=['Unnamed: 0','Unnamed: 1','Date','Cool','Funny','Useful','Author Friends','Author ID','Location','Profile URL','Profile Photo','URL'], axis=1)
o_df = o_df.rename(columns={'Author':'Date'})
o_df['Date'] = pd.to_datetime(o_df['Date']).dt.date
o_df.head()
```
