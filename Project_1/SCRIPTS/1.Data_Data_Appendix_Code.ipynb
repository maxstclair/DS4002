```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/DS4002/Project_1/DATA/final_restaurant_df.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

Numeric Variables

```python
print(df['Rating'].describe())
chart = df['Rating'].hist()
chart.set_title("Rating Distribution")
chart.set_xlabel("Rating")
chart.set_ylabel("Frequency")
plt.show()
```

```python
print(df['Author Reviews'].describe())
chart = df['Author Reviews'].hist(bins=100)
chart.set_title("Author Reviews Distribution")
chart.set_xlabel("Author Reviews")
chart.set_ylabel("Frequency")
plt.show()
```

```python
print(df['sentiment'].describe())
chart = df['sentiment'].hist()
chart.set_title("sentiment Distribution")
chart.set_xlabel("sentiment")
chart.set_ylabel("Frequency")
plt.show()
```

Categorical Variables

```python
print(df['Date'].value_counts())
print(df['Date'].value_counts().plot(kind='bar'))
```

```python
df['Date'].nunique()
```

```python
print(df['Restaurant'].value_counts())
chart = df['Restaurant'].value_counts().plot(kind='bar')
chart.set_title("Resturant Review Count")
chart.set_xlabel("Restaurant")
chart.set_ylabel("Frequency")
plt.show()
```

```python
print(df['DayOfWeek'].value_counts())
chart = df['DayOfWeek'].value_counts().plot(kind='bar')
chart.set_title("Day Of Week Count")
chart.set_xlabel("Day Of Week")
chart.set_ylabel("Frequency")
plt.show()
```
