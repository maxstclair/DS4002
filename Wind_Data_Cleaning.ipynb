This script contains code for cleaning and appending wind forecast data from 2021-2025.

```python
import pandas as pd
```

2021

```python
df_2021 = pd.read_csv('/content/miso_gen_wnd_hr_2021.csv')
df_2021 = df_2021.drop(columns=['Local Timestamp Eastern Standard Time (Interval Beginning)', 'Local Timestamp Eastern Standard Time (Interval Ending)'], axis=1)
df_2021.head()
```

2022

```python
df_2022 = pd.read_csv('/content/miso_gen_wnd_hr_2022.csv')
df_2022 = df_2022.drop(columns=['Local Timestamp Eastern Standard Time (Interval Beginning)', 'Local Timestamp Eastern Standard Time (Interval Ending)'], axis=1)
df_2022.head()
```

2023

```python
df_2023 = pd.read_csv('/content/miso_gen_wnd_hr_2023.csv')
df_2023 = df_2023.drop(columns=['Local Timestamp Eastern Standard Time (Interval Beginning)', 'Local Timestamp Eastern Standard Time (Interval Ending)'], axis=1)
df_2023.head()
```

2024

```python
df_2024 = pd.read_csv('/content/miso_gen_wnd_hr_2024.csv')
df_2024 = df_2024.drop(columns=['Local Timestamp Eastern Standard Time (Interval Beginning)', 'Local Timestamp Eastern Standard Time (Interval Ending)'], axis=1)
df_2024.head()
```

2025

```python
df_2025 = pd.read_csv('/content/miso_gen_wnd_hr_2025.csv')
df_2025 = df_2025.drop(columns=['Local Timestamp Eastern Standard Time (Interval Beginning)', 'Local Timestamp Eastern Standard Time (Interval Ending)'], axis=1)
df_2025.head()
```

Full Data Set

```python
wind_df = pd.concat([df_2021, df_2022, df_2023, df_2024, df_2025])
wind_df.head()
```

```python
wind_df.to_csv('wind.csv', index=False)
```
