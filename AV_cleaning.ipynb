```python
import pandas as pd
```

# Ten Sushi

```python
ten_df = pd.read_csv('/content/DS4002/Project_1/data/ten_sushi.csv')
```

```python
ten_df['Date'] = pd.to_datetime(ten_df['Date']).dt.date
ten_df = ten_df.drop(ten_df.columns[:2], axis=1)
ten_df = ten_df.drop(columns=['Author', 'Cool','Funny','Useful','Author Friends', '(view source)'], axis=1)
ten_df['Restaurant'] = 'Ten Sushi'
ten_df.head()
```

# Virg

```python
virg_df = pd.read_csv('/content/DS4002/Project_1/data/virginian.csv')
```
