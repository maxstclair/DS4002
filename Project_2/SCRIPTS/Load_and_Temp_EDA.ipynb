```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/DS4002/Project_2/DATA/load_temp_wind_df.csv')
```

```python
df.head()
```

Demand by Hour

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hour Number', y='MISO Total Actual Load (MW)', data=df)
plt.title('Energy Demand by Hour')
plt.xlabel('Hour')
plt.ylabel('Energy Demand')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MISO Total Wind Generation (MW)', y='SPP Total Actual Load (MW)', data=df)
plt.title('Load Demand by Wind')
plt.xlabel('Wind Generation')
plt.ylabel('Load Demand')
plt.show()
```
