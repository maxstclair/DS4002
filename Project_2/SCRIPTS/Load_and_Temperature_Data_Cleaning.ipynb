Data cleaning for hourly load demand and hourly temperature.

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/spp_load-temp_hr_2021.csv")
df2 = pd.read_csv("/content/spp_load-temp_hr_2022.csv")
df3 = pd.read_csv("/content/spp_load-temp_hr_2023.csv")
df4 = pd.read_csv("/content/spp_load-temp_hr_2024.csv")
df5 = pd.read_csv("/content/spp_load-temp_hr_2025.csv")
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df = df.drop(columns = ["SPP Total Forecast Load (MW)", "Local Timestamp Central Time (Interval Beginning)", "Local Timestamp Central Time (Interval Ending)"])
df["Local Date"] = pd.to_datetime(df["Local Date"]).dt.date
df["UTC Timestamp (Interval Ending)"] = pd.to_datetime(df["UTC Timestamp (Interval Ending)"])
df.head()
```

```python
df.to_csv('load_and_temp.csv', index=False)
```
